```python
import numpy as np
```

```python
data = np.loadtxt('./Sales.csv',delimiter = ',', skiprows = 1)
print(data)
```

```python
Total_sales_ProductA = np.sum(data[:,1])
Total_sales_ProductB= np.sum(data[:,2])
Total_sales_ProductC= np.sum(data[:,3])
print(Total_sales_ProductA,Total_sales_ProductB,Total_sales_ProductC)
```

```python
Daily_total_sales = np.sum(data[:,1:],axis=1)
print(Daily_total_sales)
```

```python
Avg_sales_ProductA = np.average(data[:,1])
Avg_sales_ProductB = np.average(data[:,2])
Avg_sales_ProductC = np.average(data[:,3])
print(Avg_sales_ProductA,Avg_sales_ProductB,Avg_sales_ProductC)
```

```python
Best_Selling = max(Total_sales_ProductA,Total_sales_ProductB,Total_sales_ProductC)
print(Best_Selling)
```

```python
Worst_day = min(Daily_total_sales)
print(Worst_day)
```

```python
Daily_growth = np.diff(Daily_total_sales)
print(Daily_growth)
```

```python
Sales_Decreasing_Days = [idx+1 for idx,val in enumerate(Daily_growth) if val<0]
print(Sales_Decreasing_Days)
```

```python
# ((today-yesterday)/(yesterday))*100 formula for daily_growth_percentage
Growth_Percentage_ProductA = ((data[1:,1] - data[:4,1])/data[:4,1])*100
Growth_Percentage_ProductB = ((data[1:,2] - data[:4,2])/data[:4,2])*100
Growth_Percentage_ProductC = ((data[1:,3] - data[:4,2])/data[:4,3])*100
print(Growth_Percentage_ProductA)
print(Growth_Percentage_ProductB)
print(Growth_Percentage_ProductC)
```

```python
Prediction_For_Next_Day = np.average(Daily_total_sales)
print(Prediction_For_Next_Day)
```

```python
Sales_Performance = ['High Sales' if x>(np.average(Daily_total_sales)) else 'Low Sales' for x in Daily_total_sales]
print(Sales_Performance)
```
